# Voters: Get Dems in a named NJ town and county

Get:
* Mountain Lakes (Morris, 1425)
* Alpine (Bergen, 0)
* Essex Fells (Essex, 0706)
* Short Hills (doesn't have its own list: is in Millburn)
* Millburn (Essex)

Results so far:
* Just 10 numbers in Alpine
* A few in Essex Fells
* Lots in Mountain Lakes

```python
import pandas as pd
import csv


# Get data headers
nametypes = njvotes.get_voteheadertypes()

njvoters = pd.read_csv('../2017 Voter data/NJvoters_allparties.csv', dtype=nametypes)
towns = ['mountain lakes', 'alpine', 'essex fells', 'millburn']
```

```python
for town in towns:
    towndems = njvoters[(njvoters['city'] == town) & (njvoters['party code'] == 'DEM')]
    towndems = towndems.drop_duplicates(keep='last', subset=['voter id'])
    towndems.to_csv('../outputs/demsin_' + town.lower() + '.csv', index=False)
    phonedems = towndems[towndems['phone number'].notnull()]
    phonedems.to_csv('../outputs/demphonesin_' + town.lower() + '.csv', index=False)
    uniquephones = towndems['phone number'].unique()
    print('{} in {}: \n{}'.format(len(uniquephones), town, uniquephones))
```
