# Voters: Get Dems in a named zipcode and county

Get:
* Mountain Lakes (Morris, 1425)
* Alpine (Bergen, 0)
* Essex Fells (Essex, 0706)
* Short Hills (doesn't have its own list: is in Millburn)
* Millburn (Essex)

Results so far:
* Just 10 numbers in Alpine
* A few in Essex Fells
* Lots in Mountain Lakes

Notes:
* City field is really quite dirty... needs cleaning (and is there somewhere we can post on cleaned data to?)

```python
import pandas as pd
import csv
import zipfile
import os
import glob
import csv 
pd.set_option("display.max_columns", 999)

county = 'ESSEX'
zipcode = '07078'

# Get data headers
fin = open('../voting data/data headers.csv', 'r')
cin = csv.reader(fin)
nameslist = cin.__next__()
nametypes = {}
for col in nameslist:
    nametypes[col] = 'str'

dfvotes = pd.read_csv('../voting data/'+county+'/ElectionHistory_cleaned.csv', dtype=nametypes)
allvoters = dfvotes[(dfvotes['zip5'] == zipcode) & (dfvotes['party code'] == 'DEM')]

dfvoters = allvoters[['voter id', 'city', 'party code', 
                        'prefix', 'first name', 'middle name', 'last name', 'suffix', 
                        'birth date', 'phone number', 'zip5']]
dfvoters.drop_duplicates(inplace=True, keep='last', subset=['voter id'])
print('Found {} dems in {}'.format(len(dfvoters), county))

dfvoters.to_csv('../outputs/demsin_' + zipcode + '.csv', index=False)
phonedems = dfvoters[dfvoters['phone number'].notnull()]
phonedems.to_csv('../outputs/demphonesin_' + zipcode + '.csv', index=False)
uniquephones = dfvoters['phone number'].unique()
print('{} in {}: \n{}'.format(len(uniquephones), zipcode, uniquephones))
```
