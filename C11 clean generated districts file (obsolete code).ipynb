# C11 clean generated districts file

```python
import csv
import pandas as pd
```

```python
infile = '../Congress11/ncjd-2011-plan-components-county-mcd.csv'
fin = open(infile, 'r')
cin = csv.reader(fin)
ignorerows = ['', 'population']
ignorefirst = ['plan:', 'plan type:', 'administrator', 'user:', 'plan components report',
               'friday, december 23, 2011']
dataset = []
district = ''
county = ''
for row in cin:
    # ignore blank rows
    rowtext = ''.join(row).strip().lower()
    # Ignore page numbers
    if rowtext.startswith('page '):
        continue
    if rowtext.startswith('plan:'):
        continue
    if rowtext.startswith('type:'):
        continue
    if rowtext in ignorerows:
        continue
    # ignore header rows
    if row[0].strip().lower() in ignorefirst:
        continue
    # Ignore continuation rows
    if '(continued)' in rowtext:
        continue
    # Ignore subtotals 
    if ' subtotal' in rowtext:
        continue
       
    # We got the row... now clean it up!
    if row[0].strip() in ['', ':']:
        row = row[1:]
    if row[0].lower().strip().startswith('district'):
        district = row[0][8:].strip()
        continue
    if 'county' in rowtext:
        county = rowtext[:-9].strip().title()
        continue
    dataset += [[district, county, row[0], ''.join(row[1:]).strip()]]
fin.close()
df = pd.DataFrame(dataset, columns=['District', 'County', 'City', 'Population'])
df.loc[df['City'].str.startswith(': '), ['City']] = df['City'].str[2:]
df.to_csv('../Congress11/ncjd-2011-districts-cleaned.csv', index=False)
df11 = df[df['District'] == '011']
df11.to_csv('../Congress11/ncjd-2011-district11-cleaned.csv', index=False)
df
```
