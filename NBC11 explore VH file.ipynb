# Inspect Kevin's copy of this dataset

```python
import pandas as pd
import njvotes


headers = ['nationbuilder_id', 'first_name', 'middle_name', 'last_name',
       'full_name', 'legal_name', 'sex', 'party', 'born_at', 'phone_number',
       'federal_district', 'state_upper_district', 'state_lower_district',
       'county_district', 'school_district', 'fire_district', 'precinct_name',
       'precinct_code', 'registered_state', 'registered_city',
       'registered_county', 'registered_zip', 'registered_address1',
       'registered_address2', 'registered_address3']
headertypes = {}
for col in headers:
    headertypes[col] = 'str'

df = pd.read_excel('../2017 Other data/Kevin/NJ Voter File 11th Congressional Modified v2.xlsx')
df
```

```python
# Clean dataset
df = df.fillna('')

citycorrections = njvotes.set_citycorrections()
df['registered_city'] = df['registered_city'].str.lower().str.strip() #Yep, there were spaces at the end of city names...
for county in citycorrections:
    for badval, goodval in citycorrections[county].items():
        df['registered_city'] = df['registered_city'].str.replace(badval, goodval)
df['registered_city'] = df['registered_city'].str.title()
df['registered_city'].value_counts()
```
