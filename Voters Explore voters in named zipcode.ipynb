# Explore Voters in Named zipcode

## Voting records

What do I want to know about someone? 
* Voter id
* Name: first, middle, last, suffix
* Phone number
* Demographics: sex, age (from birthdate), birthdate
* Voting address: street number, street name, city
* Mailing address (if different: use mailing zip code <> 07078)
* House: number of people they live with (count of address), (house value?)
* Voting record: years registered, votes for each party, years+elections (use election date+election type)

```python
import pandas as pd
import csv
pd.set_option("display.max_columns", 999)


county = 'ESSEX'
zipcode = '07078' ("Short Hills")


# Get data headers
nametypes = njvotes.get_voteheadertypes()
    
df = pd.read_csv('../2017 Voting data/NJvoters_' + county + 'allparties.csv', 
                   dtype=nametypes)
df = df[df['zip5'] == zipcode]
df = df.fillna('')
df.head()
```

```python
df = df.drop('status code',axis=1)
df = df.drop('prefix',axis=1)
df = df.drop('suffix a',axis=1)
df = df.drop('suffix b',axis=1)
df = df.drop('apt/unit no.',axis=1)
df = df.drop('address line 1',axis=1)
df = df.drop('address line 2',axis=1)
df = df.drop('city',axis=1)
df = df.drop('state',axis=1)
df = df.drop('zip4',axis=1)
df = df.drop('mailing street number',axis=1)
df = df.drop('mailing suffix a',axis=1)
df = df.drop('mailing suffix b',axis=1)
df = df.drop('mailing street name',axis=1)
df = df.drop('mailing apt/unit no.',axis=1)
df = df.drop('mailing address line 1',axis=1)
df = df.drop('mailing address line 2',axis=1)
df = df.drop('mailing city',axis=1)
df = df.drop('mailing state',axis=1)
df = df.drop('mailing country',axis=1)
df = df.drop('county precinct',axis=1)
df = df.drop('municipality',axis=1)
df = df.drop('ward',axis=1)
df = df.drop('district',axis=1)
df = df.drop('election name',axis=1)
df = df.drop('election category',axis=1)
df = df.drop('ballot type',axis=1)
df = df.drop('xx',axis=1)
df.head()
```

```python
votecounts = pd.crosstab(df['voter id'], df['party code'])
votecounts['not DEM'] = votecounts.sum(axis=1) - votecounts['DEM']
votecounts['voter id'] = votecounts.index
df.drop_duplicates(subset='voter id', inplace=True, keep='last')
votecounts.head()
```

```python
df['name'] = df['first name'] + " " + df['middle name'].str.lower() + " " + df['last name'] + df['suffix']
df = df.drop('first name',axis=1)
df = df.drop('middle name',axis=1)
df = df.drop('last name',axis=1)
df = df.drop('suffix',axis=1)
df['address'] = df['street number'] + " " + df['street name'] + ', ' + df['zip5']
df = df.drop('street number',axis=1)
df = df.drop('street name',axis=1)
df = df.drop('zip5',axis=1)
```

```python
df = pd.merge(df, votecounts[['voter id', 'DEM', 'not DEM']])
df = df[df['party code'] == 'DEM']
df
```

```python
df.rename(columns={'election date': 'last voted date', 'election type': 'last voted in',
                  'DEM' : 'Voted DEM', 'not DEM': 'Voted non-DEM'}, inplace=True)

df = df[['voter id', 'name', 'phone number', 'address', 'sex', 'birth date', 
         'date registered', 'Voted DEM', 'Voted non-DEM', 'last voted date', 
         'last voted in']]

df['last voted date'] = df['last voted date'].str[-4:]
df['date registered'] = df['date registered'].str[-4:]
df.to_csv('../outputs/shorthills_dems.csv')
df[df['phone number'] != ''].to_csv('../outputs/shorthills_demswithphones.csv')
```
