# FEC explore individual contributions files

```python
import pandas as pd
import matplotlib.pyplot as plt
import njvotes
%matplotlib inline
pd.set_option("display.max_columns", 999)


# Get voter dataset
voterheadertypes = njvotes.get_voterheadertypes()
dfvoters = pd.read_csv('../2017 Voter data/NJvoters_allparties.csv',
                       dtype=voterheadertypes) 
dfvoters = dfvoters.fillna('')
print('{}'.format(dfvoters.columns))
dfvoters
```

```python
# Get contribution dataset
headers = pd.read_csv('../2017 FEC data/indiv_header_file.csv')
df = pd.read_csv('../2017 FEC data/indiv16/itcont.txt', sep='|', 
                 names=headers.columns)
df = df.fillna('')
df['surname'] = df['NAME'].str.title().str.split(',').str[0]
df['firstnames'] = df['NAME'].str.title().str.split(',').str[1]
df['first name'] = df['firstnames'].str.split(' ').str[1]
df['zip5'] = df['ZIP_CODE'].astype(str).str[:5]

print('{}'.format(df.columns))
df
```

```python
# Pick out and prep New Jersey names
dfnj = df[df['STATE'] == 'NJ']
dfnj
```

```python
# Mix the datasets
dfall = dfnj[['EMPLOYER', 'OCCUPATION', 'TRANSACTION_PGI', 'TRANSACTION_DT',
              'TRANSACTION_AMT', 'NAME', 'first name', 'surname', 
              'zip5']].merge(dfvoters[['first name', 'last name', 'zip5',
                                       'voter id', 'party code', 
                                       'sex', 'birth date', 'phone number', 
                                       'ballot type', 'street number', 'street name',
                                       'city', 'state']], 
                            how='inner', left_on=['first name', 'surname', 'zip5'], 
                            right_on=['first name', 'last name', 'zip5'])
dfall = dfall.fillna('')
dfall
```

```python
dfall.to_csv('../2017 FEC data/2016_found_contributors.csv')
```
