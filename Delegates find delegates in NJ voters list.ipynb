```python
import pandas as pd
import csv
import zipfile
import os
import glob
import csv
```

```python
# Get list of delegates
dfpeeps = pd.read_csv('../2017 Other data/2016 DNC Delegation.csv')
dfsurs = dfpeeps['Name-Last'].str.lower().tolist()
dffl = dfpeeps['Name-First'].str.lower() + '/' + dfpeeps['Name-Last'].str.lower()
dffllist = dffl.tolist()
dfpeeps.head()
```

```python
# Find delegate names in the list of voters
nametypes = {}
for col in ['voter id', 'city', 'party code', 'prefix', 'first name', 'middle name', 
           'last name', 'suffix', 'birth date', 'phone number', 'county']:
    nametypes[col] = 'str'
njvoters = pd.read_csv('../2017 Voter data/NJvoters.csv', dtype=nametypes)
njvoters['firstlast'] = njvoters['first name'].str.lower() + '/' + \
                        njvoters['last name'].str.lower()
njvoters = njvoters.fillna('')

founddelegates = njvoters[njvoters['firstlast'].isin(dffllist)]
founddelegates.to_csv('../2017 Voter data/Delegates_found.csv')

# List the people we missed
print("Couldn't find these people: {}".format(list(set(dffllist) - 
                                                   set(founddelegates['firstlast'].unique().tolist()))))
founddelegates.head()
```
