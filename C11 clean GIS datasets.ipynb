```python
from fiona import collection


infile = '../2017 Other data/NJCD_2011_SHAPE_FILE/NJCD_2011_PLAN_SHAPE_FILE.shp'
outfile = '../2017 Other data/C11_outline.shp'
fin = collection(infile, 'r')
fout = open(outfile, 'w')
mapdata = []
for feature in fin:
    mapdata += [feature]
print('C11 features: {}'.format(mapdata[10]))
fout.write(mapdata[10])
fout.close()
```
